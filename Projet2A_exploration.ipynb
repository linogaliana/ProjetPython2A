```python
import pandas as pd 
import plotly.express as px
import plotly.graph_objects as go
import json
```

```python
path = "/Users/emanieluu/Documents/Projet_Python_2A/"

file1 = "ID_2019r.csv"
file_cap2019 = "capact_tot0019.csv"
file2 = "communes-20190101.json"
file3 = "a-dep2021.json"
# Opening JSON file
f = open(path + file2)
f2 = open(path + file3)
communes = json.load(f)
departements = json.load(f2)
```

```python
df_cap2019 = pd.read_csv(path + file_cap2019,sep = ";", encoding = "ISO-8859-1",low_memory=False)
df_cap2019.head()
```

```python
df_cap2019_2 = df_cap2019[df_cap2019["an"] == 2019]
cap_par_dep = df_cap2019_2.groupby(["dep"]).sum()["LIT"].to_frame().reset_index()
```

```python
fig = px.choropleth_mapbox(
    cap_par_dep,
    geojson= departements,
    locations='dep',
    featureidkey="properties.dep", #open-street-map
    color= cap_par_dep['LIT'],
    color_continuous_scale='YlOrRd',
    range_color=(0, max(cap_par_dep["LIT"])),
    hover_name='LIT',
    hover_data =['dep'],
    mapbox_style='open-street-map',
    zoom=4,
    center={'lat': 47, 'lon': 2},
    opacity=0.8,
    title = "Nombre de lits d'hopitalisation par départements en 2019"
   )

# fig.show()
```
